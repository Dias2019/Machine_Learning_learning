# Задача 2: аппроксимация функции

```python
### importing all the necessary libraries:

from math import sin, exp
import numpy as np
%matplotlib inline
import matplotlib.pylab as plt
from scipy import linalg
```

```python
### original function calculating values:

def func(x):
    return sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)
```

В этом задании мы будем приближать указанную функцию с помощью многочленов.

```python
### our original function without any approximation:

coordinates = np.arange(1, 15.1, 0.1)
original_function = np.array([func(coordinate) for coordinate in coordinates])
plt.plot(original_function)
plt.show()
```

## 1. Сформировать систему линейных уравнений для многочлена первой степени, который должен совпадать с функцией в точках 1 и 15

```python
### we should find out A and b in our equation (Aw = b),
### 

x1 = np.array([1 , 15])
N = 2
A1 = np.empty((0, N))
for i in range(N):
    A1_line = []
    for j in range(N):
        A1_line.append(x1[i] ** j)
    A1 = np.append(A1, np.array([A1_line]), axis = 0)

B1 = np.array([func(coordinate) for coordinate in x1])
```

```python
### solving this equation to find out w1:

w1 = linalg.solve(A1, B1)
w1
```

```python
### newly approximated function:

def func1(x):
    return w1[0] + w1[1] * x

function_1 = np.array([func1(coordinate) for coordinate in coordinates])
```

```python
### finally we should plot this function and compare it with the original one:

plt.plot(coordinates, function_1, coordinates, original_function)
plt.show()
```

Finally, we could notice that our function is a straight line which is not forming suitable approximation.

## 2. Сформировать систему линейных уравнений для многочлена второй степени, который должен совпадать с функцией в точках 1, 8 и 15

```python
### the same procedure is done as in part 1:

x2 = np.array([1 , 8, 15])
N = 3
A2 = np.empty((0, N))
for i in range(N):
    A2_line = []
    for j in range(N):
        A2_line.append(x2[i] ** j)
    A2 = np.append(A2, np.array([A2_line]), axis = 0)

B2 = np.array([func(coordinate) for coordinate in x2])
```

```python
w2 = linalg.solve(A2, B2)
w2
```

```python
### newly approximated function:

def func2(x):
    return w2[0] + w2[1] * x + w2[2] * (x ** 2)

function_2 = np.array([func2(coordinate) for coordinate in coordinates])
```

```python
### finally we should plot this function and compare it with the original one:

plt.plot(coordinates, function_1, coordinates, original_function, coordinates, function_2)
plt.show()
```

By adding more variables our function is introducing more curvature.

## 3. Сформировать систему линейных уравнений для многочлена третьей степени, который должен совпадать с функцией в точках 1, 4, 10 и 15

```python
### the same procedure is done as in part 1 and part 2:

x3 = np.array([1 , 4, 10, 15])
N = 4
A3 = np.empty((0, N))
for i in range(N):
    A3_line = []
    for j in range(N):
        A3_line.append(x3[i] ** j)
    A3 = np.append(A3, np.array([A3_line]), axis = 0)

B3 = np.array([func(coordinate) for coordinate in x3])
```

```python
### our answer for this problem:

w3 = linalg.solve(A3, B3)
w3
```

```python
### newly approximated function:

def func3(x):
    return w3[0] + w3[1] * x + w3[2] * (x ** 2) + w3[3] * (x ** 3)

function_3 = np.array([func3(coordinate) for coordinate in coordinates])
```

```python
### finally we should plot this function and compare it with the original one:

plt.plot(coordinates, function_1, coordinates, original_function, coordinates, function_2, coordinates, function_3)
plt.show()
```

In this case, our approximation function is nearly equal to our original function.
