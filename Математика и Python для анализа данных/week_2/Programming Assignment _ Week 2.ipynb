# Problem 1: Cравнение предложений

## 1. Считывание предложений из файла. Приведение к нижнему регистру.

```python
file = open("C:\\Users\\dias_\\Downloads\\sentences.txt","r")
arr = file.read().lower().split("\n")
```

```python
arr
```

## 2. Произвести токенизацию (разбиение текстов на слова). Удалить пустые слова.

```python
import re

arr_token = []
for sentence in arr:
    a = re.split('[^a-z]', sentence) ### для разбиение текстов на слова мы использовали метод split(), 
                                    ### который делить текст по каждому слову
    a = ' '.join(a).split()  ### This part is to remove empty values from formed list
    arr_token.append(a)
```

```python
### This was done to remove the empty list at the end of our list:
arr_token = [x for x in arr_token if x != []]
```

```python
arr_token
```

```python
for line in arr_token:
    print (line)
```

## 3. Составить список всех слов

```python
dict_of_words = {}

i=0
for line in arr_token:
    for word in line:
        if word not in dict_of_words:
            dict_of_words[word] = i
            i += 1
```

```python
dict_of_words
```

```python
### the number of unique words in our text:
print(len(dict_of_words))
```

## 4. Создайте матрицу размера n * d

```python
import pandas as pd

### firstly, we created a dataframe with all zeros, and with all words as columns and with all sentences as row index:
df = pd.DataFrame(dict_of_words, range(len(arr_token)))

for i in range(22):
    for j in range(254):
        df.iloc[i, j] = 0
        
df
```

```python
### filling our empty dataframe with the number of words in each sentence:
for i in range(len(arr_token)):
    for word in arr_token[i]:
        df.loc[i, word] += 1
```

```python
df
```

## 5. Найти косинусное расстояние от предложения в самой первой строке до всех остальных.

```python
from scipy.spatial.distance import cosine

dict_with_distances = {}

for i in range(22):
    distance = cosine(df.iloc[0], df.iloc[i])
    dict_key = "From 1 sentence to " + str(i+1)
    dict_with_distances[dict_key] = distance
```

```python
dict_with_distances
```

```python
min(dict_with_distances)
```

```python
list_with_distances = []
dict_with_distances_copy = dict_with_distances.copy()
for i in dict_with_distances_copy:
    list_with_distances.append(dict_with_distances_copy[i])
```

```python
### we find out the closest two values to the first sentence, then we should find out the numbers of sentences for these values 
for i in range(3):
    sentence = "Top Closest " + str(i+1)
    print(sentence, min(list_with_distances))
    list_with_distances.remove(min(list_with_distances))
```

```python
print(list(dict_with_distances.keys())[list(dict_with_distances.values()).index(0.7327387580875756)])
print(list(dict_with_distances.keys())[list(dict_with_distances.values()).index(0.7770887149698589)]) 
```
